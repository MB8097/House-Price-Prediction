# see the corr of cat feature and target

```python
def correlation_ratio(categories, measurements):
    fcat, _ = pd.factorize(categories)
    cat_num = np.max(fcat) + 1
    y_avg_array = np.zeros(cat_num)
    n_array = np.zeros(cat_num)
    
    for i in range(0, cat_num):
        cat_measures = measurements[np.argwhere(fcat == i).flatten()]
        n_array[i] = len(cat_measures)
        y_avg_array[i] = np.average(cat_measures)
    
    y_total_avg = np.sum(np.multiply(y_avg_array, n_array)) / np.sum(n_array)
    
    numerator = np.sum(np.multiply(n_array, (y_avg_array - y_total_avg) ** 2))
    denominator = np.sum((measurements - y_total_avg) ** 2)
    
    if numerator == 0:
        return 0.0
    
    return np.sqrt(numerator / denominator)

def correlation_ratios(data, categorical_features, target_feature):
    correlations = {}
    for cat_feature in categorical_features:
        ratio = correlation_ratio(data[cat_feature], data[target_feature])
        correlations[cat_feature] = ratio
    return correlations


target_feature = 'SalePrice' 


correlations = correlation_ratios(df1,cat_feat,target_feature)
sorted_correlations = sorted(correlations.items(), key=lambda item: item[1], reverse=True)

for cat_feature, ratio in sorted_correlations:
    print(f"{cat_feature}: {ratio}")
```

# now we will se the outlier

```python
def count_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return len(outliers)
outlier_count = {}
for column in df1.select_dtypes(include='int'):
    outlier_count[column] = count_outliers(df1, column)
for column, count in outlier_count.items():
    print(f'{column}: {count} outliers')
```

```python
def log_transform(df, columns):
    for column in columns:
        df[column] = np.log1p(df[column])
    return df

# Function to replace outliers with mean
def replace_outliers_with_mean(df, col):
    q1 = np.percentile(df[col], 25)
    q3 = np.percentile(df[col], 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    mean_value = df[col].mean()
    
    df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), mean_value, df[col])

# Apply appropriate outlier handling for each column
for col, num_outliers in outlier_count.items():
    if num_outliers < 30:
        replace_outliers_with_mean(df1, col)
    else:
        log_transform(df1, [col])

# Verify changes
print("Outliers handled successfully.")
```
